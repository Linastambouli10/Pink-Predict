##Preprocessing

###Key Tasks:



*   Separate features (X) and target (y).
*   Split into training and test sets (70/30 split).
*   Normalize the data (essential for models).







```python
#Importing libraries for preprocessing and splitting the data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Separate variables
X = df.drop('target',axis =1)
y = df['target']
```

```python
#Split training and test (70% training, 30% for test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Normalization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

##Modeling

1.   Logistic Regression
2.   K-Nearest Neighbors (KNN)
3.   Grid Search CV

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
log_reg= LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

##Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    print(f"{model_name}")
    #Metrics
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    #Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {model_name}')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
evaluate_model(log_reg, X_test, y_test, "Logistic Regression")
evaluate_model(knn, X_test, y_test, "KNN")
```

* Logistic Regression performs very well on this specific dataset with a 0.98 f1-score
* Disadvantages : The dataset might be small, or the model might yield False Negatives (predicting benign when it is actually malignant), which is dangerous in medicine.
Improvements : Collecting more data would be a good idea for improving this model

Last step

```python
def predict_tumor_type(tumor_features, model, scaler):
    # 1. Convert input to numpy array and reshape for a single sample
    features_array = np.array(tumor_features).reshape(1, -1)

    # 2. Normalize the features using the same scaler as training
    features_scaled = scaler.transform(features_array)

    # 3. Predict
    prediction = model.predict(features_scaled)

    # 4. Return readable result
    # Assuming 0 = Benign, 1 = Malignant
    if prediction[0] == 0:
        return "Tumeur BÃ©nigne"
    else:
        return "Tumeur Maligne"
```

```python
#Testing with an example in the dataset
test = X_test[0]
results = predict_tumor_type(test, log_reg, scaler)
print(f"Results of the diagnostics : {results}")
```
